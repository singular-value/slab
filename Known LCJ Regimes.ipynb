### Cooper Pair Box with $E_J = 0$ (No Tunnel Coupling)
Reproducing results from ยง4.2.3 of Langord Circuit QED Notes. $$\newcommand{\ket}[1]{\left|{#1}\right\rangle}$$ $$\newcommand{\bra}[1]{\left\langle{#1}\right|}$$
In this regime, the Hamiltonian only includes the diagonal KE term,
$H = E_C \sum_{N} (N-N_g)^2 \ket{N} \bra{N}$

#### 1) Good Ground State: $N_g = 0$
Ground state is well separated from next higher energy states.

```python
cpb = cq.CooperPairBox(1, 1, Ng=0)
cpb.plot(num_levels=5)
```

#### 2) Good Qubit System: $N_g = \frac{1}{2}$
Degeneracy between the first two excited states will be lifted when we add in the tunnel coupling (i.e. $E_J \neq 0$).

Note that the separation between ground energy and first excited energy is less than in the case before.

```python
cpb = cq.CooperPairBox(1, 1, 0.5)
cpb.plot(num_levels=4)
```

***
### Cooper Pair Box with $E_J << E_C$ (Small Coupling Regime)
Henceforth, we'll use $N_g = 0.5$ for gate charge offset.

Now the two excited states have different energies:

```python
cpb = cq.CooperPairBox(0.1, 1, 0.5)
cpb.plot(num_levels=4, include_wavefunctions=False)
```

In limit of $E_J \rightarrow 0$, we recover the energy degeneracy from the previous section:

```python
cpb = cq.CooperPairBox(0.04, 1, 0.5)
cpb.plot(num_levels=4, include_wavefunctions=False)
```

Reproduction of Figure 8, page 52:

```python
ground_energies, first_excited_energies, second_excited_energies = [], [], []
Ngs = [0.05 * i for i in range(0, 21)]
for Ng in Ngs:
    cpb = cq.CooperPairBox(0.1, 1, Ng)
    energies = cpb.energies(num_levels=3)
    ground_energies.append(energies[0])
    first_excited_energies.append(energies[1])
    second_excited_energies.append(energies[2])

plt.plot(Ngs, ground_energies)
plt.plot(Ngs, first_excited_energies)
plt.plot(Ngs, second_excited_energies)
plt.xlabel('Ng')
plt.ylabel('First three eigenenergies')
plt.title('Avoided Crossing in CPB')
```
